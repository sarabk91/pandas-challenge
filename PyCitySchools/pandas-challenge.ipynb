```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import numpy as np

# Read School and Student Data File and store into Pandas DataFrames
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
# Calculate the total number of unique schools
total_schools = school_data_complete['school_name'].nunique()
 
total_schools
```

# District Summary

```python
# Calculate the total number of students
total_students = school_data_complete['Student ID'].nunique()

total_students
```

```python
# Calculate the total budget
total_budget = schools_df['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = students_df['math_score'].mean()

average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = students_df['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_students) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_percentage = (students_df["reading_score"] >= 70).sum() / total_students * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_percentage = passing_math_reading_count /  float(total_students) * 100
overall_passing_percentage
```

```python
district_summary = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_percentage]
})
district_summary
```

# School Summary

```python
# Group the students data by school name
school_group = students_df.groupby('school_name')
school_group
```

```python
# Calculate the total students per school
total_students = school_group['Student ID'].count()
total_students
```

```python
# Calculate the school type (assuming it's stored in schools_df as 'type' column)
school_type = schools_df.set_index('school_name')['type']
school_type
```

```python
# Calculate the total school budget 
total_school_budget = schools_df.set_index('school_name')['budget']
total_school_budget
```

```python
# Calculate the per student budget
per_student_budget = total_school_budget / total_students
per_student_budget
```

```python
# Calculate the average math score per school
average_math_score = school_group['math_score'].mean()
average_math_score
```

```python
# Calculate the average reading score per school
average_reading_score = school_group['reading_score'].mean()
average_reading_score
```

```python
# Calculate the percentage of students passing math per school
passing_math_percentage = students_df[students_df['math_score'] >= 70].groupby('school_name')['Student ID'].count() / total_students * 100
passing_math_percentage
```

```python
# Calculate the percentage of students passing reading per school
passing_reading_percentage = students_df[students_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count() / total_students * 100
passing_reading_percentage
```

```python
# Calculate the overall passing percentage per school (passing both math and reading)
overall_passing_percentage = students_df[(students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)].groupby('school_name')['Student ID'].count() / total_students * 100
overall_passing_percentage
```

```python
# Create the school summary DataFrame
school_summary_df = pd.DataFrame({
    'School Type': school_type,
    'Total Students': total_students,
    'Total School Budget': total_school_budget,
    'Per Student Budget': per_student_budget,
    'Average Math Score': average_math_score,
    'Average Reading Score': average_reading_score,
    '% Passing Math': passing_math_percentage,
    '% Passing Reading': passing_reading_percentage,
    '% Overall Passing': overall_passing_percentage
})
school_summary_df
```

```python
# Sort the schools by % Overall Passing in descending order (highest-performing schools)
top_schools = school_summary_df.sort_values(by='% Overall Passing', ascending=False).head(5)

# Sort the schools by % Overall Passing in ascending order (lowest-performing schools)
bottom_schools = school_summary_df.sort_values(by='% Overall Passing').head(5)
bottom_schools 
```

```python
# Pivot the students data to calculate average math score per grade level and school
math_scores_by_grade = students_df.pivot_table(values='math_score', index='school_name', columns='grade', aggfunc='mean', fill_value=0)
math_scores_by_grade
```

```python
# Pivot the students data to calculate average reading score per grade level and school
reading_scores_by_grade = students_df.pivot_table(values='reading_score', index='school_name', columns='grade', aggfunc='mean', fill_value=0)
reading_scores_by_grade
```

```python
# Create bins for school size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize schools based on size bins
school_summary_df['School Size'] = pd.cut(school_summary_df['Total Students'], bins=size_bins, labels=size_labels)

# Calculate average scores and passing percentages based on school size
size_math_scores = school_summary_df.groupby('School Size')['Average Math Score'].mean()
size_reading_scores = school_summary_df.groupby('School Size')['Average Reading Score'].mean()
size_passing_math = school_summary_df.groupby('School Size')['% Passing Math'].mean()
size_passing_reading = school_summary_df.groupby('School Size')['% Passing Reading'].mean()
size_overall_passing = school_summary_df.groupby('School Size')['% Overall Passing'].mean()

# Create the size_summary DataFrame
size_summary = pd.DataFrame({
    'Average Math Score': size_math_scores,
    'Average Reading Score': size_reading_scores,
    '% Passing Math': size_passing_math,
    '% Passing Reading': size_passing_reading,
    '% Overall Passing': size_overall_passing
})

school_summary_df
```
